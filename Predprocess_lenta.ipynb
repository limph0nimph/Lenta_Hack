# Finding stable busket 

```python
def busketa(client,date):
    maska=(data['client_id']==client)&(data['chq_date']==date)
    busket=data[maska]['material'].values
    return busket
```

```python
from datetime import date
from datetime import datetime
def find_week_busket(client):
    aska=(data['client_id']==client)
    date_format = "%m/%d/%Y"
    times=pd.to_datetime(data.loc[aska,]['chq_date']).unique()
    
    #times.sort()
    stable_busket=set(busketa(client,times[0]))
    interval=0
    kk=0
    for ii in range(1,len(times)):
        time=times[ii]
        print(time)
        delta=time-times[ii-1]
        print(delta,type(delta))
        
        delta.days
        busket=busketa(client,time)
        new_interval=interval+delta.days
        if interval==0 and new_interval<7:
            interval+=delta.days
            multibusket=set(busket)
        if interval>0 and new_interval<7:
            multibusket.union(set(busket(client,time)))
            
        if new_interval>7:
            if kk==0:
                stable_busket=multibusket
            
            if kk>0:
                stable_busket.intersection(multibusket)
                    
    return stable_busket
```

```python
def find_7goes_busket(client):
    aska=(data['client_id']==client)
    times=data[aska]['chq_date'].unique()
    
    times.sort()
    mult_busket=set(busketa(client,times[0]))
    stable_busket=set(busketa(client,times[0]))
    kk=0
    for ii in range(1,len(times)-1):
        time=times[ii]
        #print(time)
      
        busket=busketa(client,time)
       
        if kk<7:
            set(mult_busket).union(busket)
        if kk==7:
            if ii<5:
                stable_busket=(mult_busket)

            else:
                stable_busket.intersection(mult_busket)
            mult_busket=busketa(client,times[ii+1])
            kk=0
        kk+=1
    return stable_busket
```

```python
def find_stable_busket(client):
    aska=(data['client_id']==client)
    times=data[aska]['chq_date'].unique()
    times.sort()
    stable_busket=set(busketa(client,times[0]))
    for ii in range(1,len(times)):
        time=times[ii]
        busket=busketa(client,time)
                    
        stable_busket=set(stable_busket).intersection(set(busket))
        
    return stable_busket
```
